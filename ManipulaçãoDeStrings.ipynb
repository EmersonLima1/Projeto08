```python
import re  # Importar o módulo re (expressões regulares)

# Exemplo de string
texto = '   Olá, mundo!   '

# Slicing: acessar substrings
substring = texto[3:8]  # Obtém a substring da posição 3 à 7
print("Substring:", substring)
print('\n') # Pular a linha

# strip(): remover espaços em branco nas extremidades
texto_sem_espacos = texto.strip()
print("Texto sem espaços em branco nas extremidades:", texto_sem_espacos)
print('\n') # Pular a linha

# lstrip() e rstrip(): remover espaços em branco no início e no final
texto_inicio_sem_espacos = texto.lstrip()
texto_final_sem_espacos = texto.rstrip()
print("Texto sem espaços em branco no início:", texto_inicio_sem_espacos)
print("Texto sem espaços em branco no final:", texto_final_sem_espacos)
print('\n') # Pular a linha

# strip() com caracteres personalizados
texto_personalizado = '###Olá, mundo!###'
texto_sem_caracteres = texto_personalizado.strip('#')
print("Texto sem caracteres personalizados:", texto_sem_caracteres)
print('\n') # Pular a linha

# removeprefix() e removesuffix(): remover prefixo e sufixo
texto_sem_prefixo = texto.removeprefix('   ')
texto_sem_sufixo = texto.removesuffix('   ')
print("Texto sem prefixo:", texto_sem_prefixo)
print("Texto sem sufixo:", texto_sem_sufixo)
print('\n') # Pular a linha

# replace(): substituir substrings
texto_substituido = texto.replace('Olá', 'Oi')
print("Texto com substituição:", texto_substituido)
print('\n') # Pular a linha

# split(): dividir a string em palavras
palavras = texto.split()
print("Palavras divididas:", palavras)
print('\n') # Pular a linha

# rsplit(): dividir a string da direita para a esquerda
palavras_direita = texto.rsplit(maxsplit=1)
print("Palavras divididas da direita para a esquerda:", palavras_direita)
print('\n') # Pular a linha

# join(): unir uma lista de strings em uma única string
lista_de_palavras = ['Olá', 'mundo', 'Python']
texto_unido = ' '.join(lista_de_palavras)
print("Texto unido:", texto_unido)
print('\n') # Pular a linha

# upper(), lower(), capitalize(): conversão de maiúsculas e minúsculas
texto_maiusculo = texto.upper()
texto_minusculo = texto.lower()
texto_capitalizado = texto.capitalize()
print("Texto em maiúsculas:", texto_maiusculo)
print("Texto em minúsculas:", texto_minusculo)
print("Texto capitalizado:", texto_capitalizado)
print('\n') # Pular a linha

# islower(), isupper(): verifica se a string é totalmente maiúscula ou minúscula
print("A string está em minúsculas?", texto.islower())
print("A string está em maiúsculas?", texto.isupper())
print('\n') # Pular a linha

# isalpha(), isnumeric(), isalnum(): verifica tipos de caracteres
texto1 = 'Python'
texto2 = '123'
texto3 = 'Python123'
texto4 = 'Python-123'
print("É alfabético?", texto1.isalpha(), texto2.isnumeric(), texto3.isalnum())
print("É alfabético?", texto4.isalpha(), texto4.isnumeric(), texto4.isalnum())
print('\n') # Pular a linha

# count(): contar ocorrências de substring
contagem = texto.count('o')
print("Número de 'o's na string:", contagem)
print('\n') # Pular a linha

# find(): encontrar a posição da primeira ocorrência de substring
posicao = texto.find('mundo')
print("Posição da primeira ocorrência de 'mundo':", posicao)
print('\n') # Pular a linha

# rfind(): encontrar a posição da última ocorrência de substring
ultima_posicao = texto.rfind('o')
print("Posição da última ocorrência de 'o':", ultima_posicao)
print('\n') # Pular a linha

# startswith() e endswith(): verificar prefixo e sufixo
comeca_com_espaco = texto.startswith(' ')
termina_com_espaco = texto.endswith(' ')
print("Começa com espaço?", comeca_com_espaco)
print("Termina com espaço?", termina_com_espaco)
print('\n') # Pular a linha

# partition(): dividir a string em três partes com base na primeira ocorrência
partes = texto.partition(',')
print("Partes divididas:", partes)
print('\n') # Pular a linha

# center(), ljust(), rjust(): justificar texto
texto_centralizado = texto.center(30, '-')
texto_esquerda = texto.ljust(30, '-')
texto_direita = texto.rjust(30, '-')
print("Texto centralizado:", texto_centralizado)
print("Texto à esquerda:", texto_esquerda)
print("Texto à direita:", texto_direita)
print('\n') # Pular a linha

# f-Strings: formatação de strings (Python 3.6+)
numero = 42
linguagem = 'Python'
frase_formatada = f'{linguagem} é o número {numero} em programação!'
print("Frase formatada:", frase_formatada)
print('\n') # Pular a linha

# swapcase(): alternar maiúsculas e minúsculas
texto_alternado = texto.swapcase()
print("Texto com alternância de maiúsculas e minúsculas:", texto_alternado)
print('\n') # Pular a linha

# zfill(): preencher com zeros à esquerda
numero_zfill = '42'.zfill(7) # Nesse exemplo, 7 é o número total de caracteres que queremos que tenha na string resultante, logo adicionará apenas 5 zeros a esquerda
print("Número preenchido com zeros à esquerda:", numero_zfill)
print('\n') # Pular a linha

# Exemplo de string
texto = 'string    métodos em python'

# re.sub(): substituir um padrão específico usando expressões regulares
padrao = "\s+"  # Padrão para corresponder a espaços em branco
substituicao = "-"  # Substituir espaços em branco por traços
texto_substituido = re.sub(padrao, substituicao, texto)
print("Texto com substituição usando expressões regulares:", texto_substituido)
```
